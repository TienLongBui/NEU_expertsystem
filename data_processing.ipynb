### 2.2 Column imputation
This section will rename Vietnamese columns into English, adding a new column according to the surveyers

```python
# Remove STT column
students_df.drop(columns="STT", inplace=True)
lecturers_df.drop(columns="Stt", inplace=True)
staff_df.drop(columns="STT", inplace=True)
```

```python
# Encoding column name
# students_df
students_df.rename(columns={
    "Trường Đại học": "University",
    "Giới tính": "Gender",
    "Sinh viên năm": "Student_year",
    "Khu vực": "Region"
}, inplace=True)
students_df["Group"] = "Student"

# lecturers_df
lecturers_df.rename(columns={
    "Trường Đại học": "University",
    "Giới tính": "Gender",
    "Độ tuổi": "Age",
    "Số năm trong lĩnh vực giảng dạy": "Years_of_Experience",
    "Học hàm/học vị": "Academic_Title"
}, inplace=True)
lecturers_df["Group"] = "Lecturer"

# staff_df
staff_df.rename(columns={
    "Trường Đại học": "University",
    "Giới tính": "Gender",
    "Độ tuổi": "Age",
    "Số năm trong lĩnh vực giảng dạy": "Years_of_Experience",
    "Học hàm/học vị": "Academic_Title",
    "Ngạch thầy/cô": "Position"
}, inplace=True)
staff_df["Group"] = "Staff"
```

```python
students_df
```

### 2.3 Encode column university and region
This section will map the university's name to number label

```python
print(students_df['University'].unique())
print(staff_df['University'].unique())
print(lecturers_df['University'].unique())

print(students_df['Region'].unique())
```

```python
# Remove blank space
students_df['University'] = students_df['University'].str.strip()
staff_df['University'] = staff_df['University'].str.strip()
lecturers_df['University'] = lecturers_df['University'].str.strip()
```

```python
# Dictionary mapping
university_mapping = {
    'Đại học Kinh tế Quốc dân': 1,
    'Đại học Ngoại thương': 2,
    'Học viện Tài chính': 3,
    'Học viện Ngân hàng': 4,
    'Đại học Thương mại': 5,
}

region_mapping = {
    'KV1': 1,
    'KV2': 2,
    'KV2-NT': 3,
    'KV3': 4
}
```

```python
# Encode 'University' column 
students_df['University'] = students_df['University'].map(university_mapping)
staff_df['University'] = staff_df['University'].map(university_mapping)
lecturers_df['University'] = lecturers_df['University'].map(university_mapping)

# Encode 'Region' column 
students_df['Region'] = students_df['Region'].map(region_mapping)
```

```python
print(students_df['University'].unique())
print(staff_df['University'].unique())
print(lecturers_df['University'].unique())

print(students_df['Region'].unique())
```

### 2.4. Merging all dataframe
- This section will merge 3 dataframe into single 1 dataframe for future implementation. 
- The merging dataframe name **combined_df**

```python
# Merging dataframe
combined_df = pd.concat([students_df, lecturers_df, staff_df], ignore_index=True)
```

```python
combined_df
```

```python
# Filling "Not Applicable" to demographic observations
combined_df.fillna({
    "Year": "Not Applicable",
    "Region": "Not Applicable",
    "Age": 0,
    "Years_of_Experience": 0,
    "Academic_Title": "Not Applicable",
    "Position": "Not Applicable"
}, inplace=True)
```

```python
#combined_df
```

```python
column_order = [
    "Group","University", "Gender", "Student_year", "Region", "Age", "Years_of_Experience", "Academic_Title", "Position",
    "PE1", "PE2", "PE3", "PE4", "PE5", "PE6",
    "EE1", "EE2", "EE3", "EE4", "EE5",
    "SI1", "SI2", "SI3", "SI4", "SI5",
    "FC1", "FC2", "FC3", "FC4", "FC5",
    "TAC1", "TAC2", "TAC3", "TAC4", "TAC5",
    "TEC1", "TEC2", "TEC3", "TEC4", "TEC5",
    "TTF1", "TTF2", "TTF3",
    "PS1", "PS2", "PS3",
    "BI1", "BI2", "BI3",
    "UB1", "UB2", "UB3"
]

# Rearrange columns
combined_df = combined_df[column_order]
print(combined_df.head())
```

## 3. Exploratory data analysis [EDA]

```python
# View all columns
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

print(combined_df.describe())
```

```python
print(combined_df["Group"].value_counts())
```

```python
# Histogram for numeric
numeric_columns = combined_df.select_dtypes(include=['int64', 'float64']).columns
for col in numeric_columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(combined_df[col], kde=True, bins=30, color="blue")
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

```python
# Counting categorical (Group)
categorical_columns = combined_df.select_dtypes(include=['object']).columns
for col in categorical_columns:
    plt.figure(figsize=(8, 5))
    sns.countplot(data=combined_df, x=col, palette="viridis")
    plt.title(f"Count of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

```python
# Filter numeric column
numeric_df = combined_df.select_dtypes(include=["int64", "float64"])

# Correlation matrix
correlation_matrix = numeric_df.corr()

# Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
# Distribution by group
for col in numeric_columns:
    plt.figure(figsize=(8, 5))
    sns.boxplot(data=combined_df, x="Group", y=col, palette="Set3")
    plt.title(f"Distribution of {col} by Group")
    plt.xlabel("Group")
    plt.ylabel(col)
    plt.show()
```

```python
# Distribution of gender in each group
plt.figure(figsize=(8, 5))
sns.countplot(data=combined_df, x="Gender", hue="Group", palette="pastel")
plt.title("Gender Distribution by Group")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(title="Group")
plt.show()
```
