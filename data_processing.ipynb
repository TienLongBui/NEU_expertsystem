## Configuration

```python
# Library configuration
#pip3 install openpyxl

import pandas as pd
import numpy as np
```

```python
# Data preparation
file_path = "data_nckh.xlsx" 
data = pd.ExcelFile(file_path)
```

```python
# Printing sheet names
print("Sheets name:", data.sheet_names)
```

```python
# Splitting sheets into different dataframes
students_df = pd.read_excel(data, sheet_name="SV578")
lecturers_df = pd.read_excel(data, sheet_name="GV120")
staff_df = pd.read_excel(data, sheet_name="CB94")
```

```python
# Remove STT column
students_df.drop(columns="STT", inplace=True)
lecturers_df.drop(columns="Stt", inplace=True)
staff_df.drop(columns="STT", inplace=True)
```

```python
# Encoding column name
# students_df
students_df.rename(columns={
    "Trường Đại học": "University",
    "Giới tính": "Gender",
    "Sinh viên năm": "Year",
    "Khu vực": "Region"
}, inplace=True)
students_df["Group"] = "Student"

# lecturers_df
lecturers_df.rename(columns={
    "Trường Đại học": "University",
    "Giới tính": "Gender",
    "Độ tuổi": "Age",
    "Số năm trong lĩnh vực giảng dạy": "Years_of_Experience",
    "Học hàm/học vị": "Academic_Title"
}, inplace=True)
lecturers_df["Group"] = "Lecturer"

# staff_df
staff_df.rename(columns={
    "Trường Đại học": "University",
    "Giới tính": "Gender",
    "Độ tuổi": "Age",
    "Số năm trong lĩnh vực giảng dạy": "Years_of_Experience",
    "Học hàm/học vị": "Academic_Title",
    "Ngạch thầy/cô": "Position"
}, inplace=True)
staff_df["Group"] = "Staff"
```
